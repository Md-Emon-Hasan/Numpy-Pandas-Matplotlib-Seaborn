```python
import pandas as pd
import matplotlib.pyplot as plt
gas = pd.read_csv('gas_prices.csv')
plt.figure(figsize=(8,5))
plt.title('Gas Price over time (in USD)')
for country in gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],marker='.')
print(gas.Year[::3])
plt.xticks(gas.Year[::3])
plt.ylabel('Year')
plt.ylabel('US Dollers')
plt.legend()
plt.show()
```

```python
gas = pd.read_csv('gas_prices.csv')
plt.figure(figsize=(8,5))
plt.title('Gas Price over tiem (in USD)')
country_to_look_at = ['Australia','USA','Canada','South korea']
for country in gas:
    if country in country_to_look_at:
        plt.plot(gas.Year,gas[country],marker='.')
print(gas.Year[::3])
plt.xticks(gas.Year[::3])
plt.xlabel('Year')
plt.ylabel('US Dollers')
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
fifa = pd.read_csv('fifa_data.csv')
plt.figure(figsize=(5,8))
barcelona = fifa.loc[fifa.Club == "FC Barcelona"]['Overall']
madrid = fifa.loc[fifa.Club == "Real Madrid"]['Overall']
revs = fifa.loc[fifa.Club == 'New England Revolution']['Overall']
lebels = ['FC Barcelona','Real Madrid','New England Revolution']
plt.title('Professional soccer Team Comparison')
plt.ylabel('FIFA Overall Rating')
boxes = plt.boxplot([barcelona, madrid, revs],labels=lebels,patch_artist=True)
for box in boxes['boxes']:
    # set edge color
    box.set(color='lime',linewidth=2)
    # change fill color
    box.set(facecolor='green')
plt.show()
```
